```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Create data
#           BERT    Roberta     DistilBERT  DeBERTa     mBERT       XLMRoBERTa  HiRoBERTa   BiLSTM      MNB     RF      SVM
ta16_30K = [0.8738,	0.8667,	    0.9133,	    0.8875,	    0.8196,	    0.9046,	    0.8067,     0.8512,     0.82,   0.8433, 0.875   ]
ta20_30K = [0.7883,	0.837,	    0.8703, 	0.8663, 	0.707,  	0.867,	    0.7153, 	0.822,  	0.731,	0.767,	0.8086  ]
x = np.arange(len(ta16_30K))
width = 0.2
```

```python
# Index of bars for different models (to assign colors)
mono_tm_idx = [0, 1, 2, 3]
multi_tm_idx = [4, 5, 6]
rnn_idx = [7]
ml_idx = [8, 9, 10]

plt.figure(figsize=(18, 6), dpi=150)

# Plot grouped bar chart
ta16_bar = plt.bar(x-width/2, ta16_30K, width, alpha=1)
ta20_bar = plt.bar(x+width*1.1/2, ta20_30K, width, alpha=1)

for idx in mono_tm_idx:
    ta16_bar[idx].set_color('slategrey')
    ta20_bar[idx].set_color('lightsteelblue')

for idx in multi_tm_idx:
    ta16_bar[idx].set_color('teal')
    ta20_bar[idx].set_color('lightseagreen')

for idx in rnn_idx:
    ta16_bar[idx].set_color('goldenrod')
    ta20_bar[idx].set_color('navajowhite')

for idx in ml_idx:
    ta16_bar[idx].set_color('indianred')
    ta20_bar[idx].set_color('salmon')

plt.xticks(x, ['BERT', 'RoBERTa', 'DistilBERT', 'DeBERTa', 'mBERT', 'XLMRoBERTa', 'HindiRoBERTa', 'Bi-LSTM', 'MNB', 'RF', 'SVM'])
plt.xlabel("Models", labelpad=20)
plt.ylabel("Test Accuracy (%)", labelpad=20)
plt.ylim(bottom=0.5)
plt.grid(True, axis='y')
# plt.legend(["16-classes", "20-classes"], loc='best')
plt.title('Dataset Size: 30000 (Large)', pad=15)

# Try saving in SVG (vector format)
plt.savefig('/home/suyogyat/research/nepali-transformers-evaluation/code/results/accuracy_by_class_v1.svg', format='svg', dpi=300)

plt.show()
```
